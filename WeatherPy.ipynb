```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = 'output_data/cities.csv'

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#setup api query variables
url = 'http://api.openweathermap.org/data/2.5/weather?'
units = 'Imperial'
query_url = f'{url}appid={api_key}&q='

#define lists to be populated with data from api
city_count = 0
date = []
lat = []
max_temp = []
humidity = []
cloudiness = []
wind_speed = []

#run api through loop to amend list once data retrieved and handle exceptions
print(f'beginning data retrieval')
print('-'*25)

for city in cities:
    city_count += 1
    print(f'processing record {city_count} | {city}')
    try:
        response = requests.get(f'{query_url}{city}&units={units}').json()
    except ValueError:
        print('city not found. skipping...')
    try:
        date.append(time.strftime('%m-%d-%Y', time.localtime(response['dt'])))
    except KeyError:
        date.append(None)
    try:
        lat.append(response['coord']['lat'])
    except KeyError:
        lat.append(None)
    try:
        max_temp.append(response['main']['temp_max'])
    except KeyError:
        max_temp.append(None)
    try:
        humidity.append(response['main']['humidity'])
    except KeyError:
        humidity.append(None)    
    try:
        cloudiness.append(response['clouds']['all'])
    except KeyError:
        cloudiness.append(None)    
    try:
        wind_speed.append(response['wind']['speed'])
    except KeyError:
        wind_speed.append(None)

print(f'-'*25)
print(f'data retrieval complete')
print(f'-'*25)
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#create dictionary with lists populated from api loop
weather_dict = {'city': cities, 'latitude': lat, 'date': date, 'max temp': max_temp, 'humidity': humidity, 'cloudiness': cloudiness, 'wind speed': wind_speed}

#create dataframe from dictionary
weather_df = pd.DataFrame(weather_dict)

#clean dataframe of null values from exception handling
weather_df_clean = weather_df.dropna(how='any')

#check dataframe to make sure not remaining null values
weather_df_clean.isnull().any()
```

```python
#check count of dataframe
weather_df_clean.count()
```

```python
#define date api loop was run
run_date = max(weather_df_clean['date'])

#output dataframe to csv
weather_df_clean.to_csv(output_data_file)

#display dataframe
weather_df_clean.head()
```

### Plotting the Data

#### Latitude vs. Temperature Plot

```python
#build scatter plot
plt.scatter(weather_df_clean['latitude'] , weather_df_clean['max temp'], c = 'steelblue', alpha = .40, linewidths = 1, edgecolors = 'black')

#define x and y limits for axis
plt.xlim(min(weather_df_clean['latitude'])*1.15,max(weather_df_clean['latitude'])*1.10)
plt.ylim(min(weather_df_clean['max temp'])*1.05,max( weather_df_clean['max temp'])*1.05)

#label scatter plot
plt.suptitle(f'city latitude vs max temperature ({run_date})', fontsize = 12)
plt.xlabel('latitude', fontsize = 10)
plt.ylabel('max temperature (F)', fontsize = 10)
plt.grid(True, axis = 'both')

#save figure
plt.savefig('latvstemp.png')

#show plot
plt.show()
```
